```python
import pandas as pd
import matplotlib.pyplot as plt

# Replicate 1
cmp2_rep1_area = pd.read_csv('cmp2_rep1_area_noTermLoops.xvg', delim_whitespace=True, comment='@', header=None)
cmp2_rep1_gyrate = pd.read_csv('cmp2_rep1_gyrate_noTermLoops.xvg', delim_whitespace=True, comment='@', header=None)
cmp2_rep1_hydrogen = pd.read_csv('cmp2_rep1_hydrogen_noTermLoops.xvg', delim_whitespace=True, comment='@', header=None)
cmp2_rep1_rmsd_lig = pd.read_csv('cmp2_rep1_rmsd_lig.xvg', delim_whitespace=True, comment='@', header=None)
cmp2_rep1_rmsf = pd.read_csv('cmp2_rep1_rmsf_noTermLoops.xvg', delim_whitespace=True, comment='@', header=None)
cmp2_rep1_rmsd = pd.read_csv('cmp2_rep1_rmsd_noTermLoops.xvg', delim_whitespace=True, comment='@', header=None)

# Replicate 2
cmp2_rep2_area = pd.read_csv('cmp2_rep2_area_noTermLoops.xvg', delim_whitespace=True, comment='@', header=None)
cmp2_rep2_gyrate = pd.read_csv('cmp2_rep2_gyrate_noTermLoops.xvg', delim_whitespace=True, comment='@', header=None)
cmp2_rep2_hydrogen = pd.read_csv('cmp2_rep2_hydrogen_noTermLoops.xvg', delim_whitespace=True, comment='@', header=None)
cmp2_rep2_rmsd_lig = pd.read_csv('cmp2_rep2_rmsd_lig.xvg', delim_whitespace=True, comment='@', header=None)
cmp2_rep2_rmsf = pd.read_csv('cmp2_rep2_rmsf_noTermLoops.xvg', delim_whitespace=True, comment='@', header=None)
cmp2_rep2_rmsd = pd.read_csv('cmp2_rep2_rmsd_noTermLoops.xvg', delim_whitespace=True, comment='@', header=None)

# Replicate 3
cmp2_rep3_area = pd.read_csv('cmp2_rep3_area_noTermLoops.xvg', delim_whitespace=True, comment='@', header=None)
cmp2_rep3_gyrate = pd.read_csv('cmp2_rep3_gyrate_noTermLoops.xvg', delim_whitespace=True, comment='@', header=None)
cmp2_rep3_hydrogen = pd.read_csv('cmp2_rep3_hydrogen_noTermLoops.xvg', delim_whitespace=True, comment='@', header=None)
cmp2_rep3_rmsd_lig = pd.read_csv('cmp2_rep3_rmsd_lig.xvg', delim_whitespace=True, comment='@', header=None)
cmp2_rep3_rmsf = pd.read_csv('cmp2_rep3_rmsf_noTermLoops.xvg', delim_whitespace=True, comment='@', header=None)
cmp2_rep3_rmsd = pd.read_csv('cmp2_rep3_rmsd_noTermLoops.xvg', delim_whitespace=True, comment='@', header=None)
```

```python
from matplotlib.ticker import FormatStrFormatter

def plot_replicates(data1, data2, data3, title, xlabel, ylabel, legend_labels, output_file, divide_x_by_1000=False):
    plt.figure(figsize=(10, 6))
    
    # Divide x-axis values by 1000 if specified
    if divide_x_by_1000:
        plt.plot(data1[0] / 1000, data1[1], label=legend_labels[0])
        plt.plot(data2[0] / 1000, data2[1], label=legend_labels[1])
        plt.plot(data3[0] / 1000, data3[1], label=legend_labels[2])
        plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
    else:
        plt.plot(data1[0], data1[1], label=legend_labels[0])
        plt.plot(data2[0], data2[1], label=legend_labels[1])
        plt.plot(data3[0], data3[1], label=legend_labels[2])
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Set the legend outside the chart
    plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust layout to make space for legend
    plt.grid(True)
    plt.savefig(output_file)
    plt.show()

# Plot Area
plot_replicates(cmp2_rep1_area, cmp2_rep2_area, cmp2_rep3_area,
                'SASA', 'Time (ns)', 'Area (nm²)',
                ['Replicate 1', 'Replicate 2', 'Replicate 3'],
                'cmp2_area.png')

# Plot Gyrate
plot_replicates(cmp2_rep1_gyrate, cmp2_rep2_gyrate, cmp2_rep3_gyrate,
                'Radius of Gyration', 'Time (ns)', 'Gyration (nm)',
                ['Replicate 1', 'Replicate 2', 'Replicate 3'],
                'cmp2_gyrate.png', divide_x_by_1000=True)

# Plot Hydrogen Bonds
plot_replicates(cmp2_rep1_hydrogen, cmp2_rep2_hydrogen, cmp2_rep3_hydrogen,
                'Hydrogen Bonds', 'Time (ns)', 'Hbonds',
                ['Replicate 1', 'Replicate 2', 'Replicate 3'],
                'cmp2_hydrogen.png')

# Plot RMSD (Ligand)
plot_replicates(cmp2_rep1_rmsd_lig, cmp2_rep2_rmsd_lig, cmp2_rep3_rmsd_lig,
                'RMSD (Ligand)', 'Time (ns)', 'RMSD (nm)',
                ['Replicate 1', 'Replicate 2', 'Replicate 3'],
                'cmp2_rmsd_lig.png')

# Plot RMSF
plot_replicates(cmp2_rep1_rmsf, cmp2_rep2_rmsf, cmp2_rep3_rmsf,
                'RMSF', 'Residue', 'RMSF (nm)',
                ['Replicate 1', 'Replicate 2', 'Replicate 3'],
                'cmp2_rmsf.png')

# Plot RMSD (Protein)
plot_replicates(cmp2_rep1_rmsd, cmp2_rep2_rmsd, cmp2_rep3_rmsd,
                'RMSD', 'Time (ns)', 'RMSD (nm)',
                ['Replicate 1', 'Replicate 2', 'Replicate 3'],
                'cmp2_rmsd.png')
```
